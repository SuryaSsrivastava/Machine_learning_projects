## KNN ALGORITHM

```python
def dist(x1,x2):
    return np.sqrt(sum((x1-x2)**2))

def knn(X_train,Y_train,X_test,Y_test,k=5):
    distance=[]
    m=X_test.shape[0]
    for i in range(m):
        d=dist(X_test[i,:],X_train[i,:])
        distance.append((d,Y_train[i]))
    distance=sorted(distance)
    val=distance[:k]
    print(val)
```

```python
knn(X_train,Y_train,X_test,Y_test)
```

## inbuilt KNN

```python
from sklearn import neighbors
from sklearn import metrics
model = neighbors.KNeighborsClassifier(n_neighbors=3)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split( 
              X, Y, test_size = 0.2, random_state = 0)
Y_train=Y_train.astype('int')
Y_test=Y_test.astype('int')
```

```python
model.fit(X_train,Y_train)
```

```python
predict = model.predict(X_test)
print(predict)
```

```python
from sklearn.metrics import accuracy_score #for checking the model accuracy
print('The accuracy of the KNN is',metrics.accuracy_score(predict,Y_test))
```
